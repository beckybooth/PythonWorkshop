#### Array operations

Simple mathematical operations on Numpy arrays are element-wise. This means that the operation is done to each element individually. 

<font color="#33A1DE">
#### Example:

```python
x = np.random.randint(20, size=5)
print('x =', x , '\n')

print('x + 2 =' , x + 2)
print('x - 2 =' , x - 2)
print('x * 2 =' , x * 2)
print('x / 2 =' , x / 2)
print('x ** 2 =' , x ** 2)
```

### Passing a subplot axis to a function

In Python, everything is an object, including plots, and any object can passed to functions as arguments. This allows us to write functions to label and customize our subplots, to save time and prevent writing the same code over and over.

<font color="#33A1DE">
#### Example 5: A function to plot, add a grid, and set the title of a subplot

```python
def plot_grid_and_title(ax, x, y, title, grid=True):
    # note that the default value of grid is set to True
    # so if no value is given, grid will be True
    # but the user can choose to set grid=False in the function call
    ax.plot(x,y)
    
    ax.set_title(title)
    
    if grid:
        ax.grid('on')

# make the data to plot
x = np.linspace(-2,2, 20)
y1 = x**2
y2 = x**3

# create the figure and subplots
fig, ax = plt.subplots(1,2, figsize=(10,4))

# call plot_grid_and_title to plot and label the data
plot_grid_and_title(ax[0], x, y1, '$y=x^2$')
plot_grid_and_title(ax[1], x, y2, '$y=x^3$', grid=False)

plt.tight_layout()
```

#### Example 6: Generating histograms

```python
# --------------------- create some data to plot --------------------- 

# this produces an array containing 5000 floats between 0 and 1
# sampled from a uniform distribution
x = np.random.random(5000) 

# ------------------------- labeling function ------------------------ 

def label_hist(ax, title, xlabel, ylabel, grid=True):
    ax.set_title(title, fontsize=13)
    ax.set_xlabel(xlabel, fontsize=13)
    ax.set_ylabel(ylabel, fontsize=13)
    
    if grid: ax.grid('on')

# ------------------------------ plotting ---------------------------- 

fig, ax = plt.subplots(1,3, figsize=(15,4))

# left
ax[0].hist(x);
label_hist(ax[0], 'Default histogram', 'number', 'counts')

# middle
ax[1].hist(x, bins=25, facecolor='#FFC1C1', edgecolor='black');
label_hist(ax[1], 'Customized histogram', 'number', 'counts')

# right
ax[2].hist(x, bins=10, facecolor='#BFEFFF', edgecolor='black', normed=True);
label_hist(ax[2], 'Normalized histogram', 'number', 'counts')

plt.tight_layout()
```

### Task 3.3: Fitting a model to data (was good for a Nobel prize)
<font color = "black"><br>
    
*Adapted from the exercise written by Gijs Nelemans for Programmeren 2 (Introductory Scientific Programming with Python) at Radboud University Nijmegen* 

From the theory of General Relativity it can be derived that the global evolution of the Universe (described by the scale factor of the Universe R) is governed by the Friedmann equation

$$\bigg(\frac{dR}{dt}\bigg) = \frac{8\pi G\rho}{3} - \frac{kc^2}{R^2}$$

with $\rho$ the density (or density equivalent with the energy density via $E=mc^2$) and $k$ a constant that depends on the geometry of space and in our Universe is (very close to) 0. It can be interpreted in a Newtonian fashion as a kinetic energy term on the left hand side and a potential energy term due to gravity on the right hand side (similarly to a ball thrown upwards on Earth). From that interpretation it is clear that the expansion is expected to decelerate (as the ball decelerates in the Earth gravitational field).

The expansion in the past and thus the deceleration can be measured directly by looking at the scale factor as a function of lookback time (i.e. distance, since light has a finite velocity and thus objects far away are seen at early times). This can be done by comparing the physical distance to an object with the redshift of the light it emits (which is caused by an increase of $R$.

In this exercise we will use supernovae to measure the deceleration of the Universe, using data from the paper “Observational Evidence from Supernovae for an Accelerating Universe and a Cosmological Constant” (Astronomical Journal, 116, 1009, 1998). Two data files containing tables (z_mu.dat and logcz_mu.dat) have been provided, or if you prefer, you can get them directly from the website of the journal ( http://iopscience.iop.org/1538-3881/116/3/1009/)

<font color="blue">**(a)**<font color="black"> Use genfromtxt as shown above to read in the in the redshift $z = v/c$, the so-called distance modulus $\mu$ and its error from the first file (z_mu.dat). In the second file the redshift is given as $\log cz$, i.e. the log of the speed of light times the redshift, but note that the unit of $cz$ that is used is $km/s$ not $m/s$! Convert the $\log cz$ array to a $z$ array.

<font color="blue">**(b)**<font color="black"> Collect the data from both files into a set of three arrays, $z$, $\mu$ and error, or into a single 3D array.

<font color="blue">**(c)**<font color="black"> The distance modulus is an archaic unit used by astronomers and denotes the number of “magnitudes” that an object looks fainter than the same object at a reference distance of $10pc$. Magnitudes originate in the past when they were used to estimate the brightness of objects by eye. They can be converted to the more physical unit of “luminosity distance” $D_L$ by

$$\mu = 5\log D_L +25$$

where $D_L$ has the unit $Mpc$. Calculate $D_L$ and its error from the values of $\mu$. 

(OPTIONAL: plot the so-called Hubble diagram in which $\mu$ is plotted versus $z$, with errorbars!)

<font color="blue">**(d)**<font color="black"> The solution for $R(t)$ can be locally (i.e. around the current time) be written as

$$R(t) = R_0 + \dot{R}t + \frac{1}{2}\ddot{R}t^2$$

With definition of the Hubble constant $H_0 = \dot{R}/R_0$ and the “deceleration parameter” $q_0 = −\ddot{R}/(R_0H_0^2)$

$$\frac{R(t)}{R_0} = 1 + H_0t - 0.5H_0^2 q_0 t^2$$

With some algebra it can be shown that this yields (for $D_L$ in $Mpc$ and $c$ in $km/s$!)

$$D_L(z) = \frac{c}{H_0}[z+0.5(1-q_0)z^2]$$

Fit this function to the data and thus determine the values of $H_0$ and $q_0$. The minus sign in the definition of $q_0$ reflects the assumption that the expansion of the Universe should be slowing down, yielding a positive value of $q_0$. The data you just fitted showed this to be incorrect and for that finding some of the authors of this paper were awarded the 2011 Nobel Prize in Physics.

### Importing FITS images

The astropy module provides tools that can be used to import FITS (flexible image transport system) files as Numpy arrays.

```python
# import the fits module from astropy
from astropy.io import fits

# open a sample fits file
examplefits = fits.open('502nmos.fits')
# A FITS file taken with the NASA/ESA Hubble Space Telescope WFPC2 camera of the 
# star-forming nebula Messier 17 through a narrow-band 502 nm filter ([OIII]).
# downloaded from https://www.spacetelescope.org/projects/fits_liberator/datasets_archives/

Messier17 = examplefits[0].data

# uncomment below to read the fits header
#print(examplefits[0].header)
```

The image is loaded as a Numpy array, we can display it using imshow.

```python
f, ax = plt.subplots(1,3, figsize=(15,4))

im1 = ax[0].imshow(Messier17, vmin=0, vmax=500)
ax[0].set_title('default colormap')
plt.colorbar(im1, ax=ax[0])


im2 = ax[1].imshow(Messier17, vmin=0, vmax=500, cmap='gray')
ax[1].set_title('gray colormap')
plt.colorbar(im2, ax=ax[1])

im3 = ax[2].imshow(Messier17, vmin=0, vmax=500, cmap='gray_r')
ax[2].set_title('gray colormap reversed')
plt.colorbar(im3, ax=ax[2]);
```
